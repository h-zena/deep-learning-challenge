```python
# Step 1: Import dependencies and read the data
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
# Read the data
data = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
```

```python

# Step 2: Preprocess the dataset
# Drop irrelevant columns
data = data.drop(columns=['EIN', 'NAME'])

# Binning for 'APPLICATION_TYPE' column
application_counts = data['APPLICATION_TYPE'].value_counts()
replace_application = list(application_counts[application_counts < 500].index)

for app in replace_application:
    data['APPLICATION_TYPE'] = data['APPLICATION_TYPE'].replace(app, 'Other')

# Binning for 'CLASSIFICATION' column
classification_counts = data['CLASSIFICATION'].value_counts()
replace_class = list(classification_counts[classification_counts < 1000].index)

for cls in replace_class:
    data['CLASSIFICATION'] = data['CLASSIFICATION'].replace(cls, 'Other')

# Convert categorical data to numeric using one-hot encoding
data_encoded = pd.get_dummies(data)

# Split the data into features and target
X = data_encoded.drop(columns=['IS_SUCCESSFUL'])
y = data_encoded['IS_SUCCESSFUL']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

```

```python
# Step 3: Design a neural network model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=80, activation='relu', input_dim=len(X_train.columns)),
    tf.keras.layers.Dense(units=30, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])
```

```python
# Step 4: Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Step 5: Train the model
model.fit(X_train, y_train, epochs=100, verbose=1)
```

```python
# Step 6: Evaluate the model
model_loss, model_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
